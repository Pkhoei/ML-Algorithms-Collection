```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,3,7,18,1,3,9,11]
plt.plot(x,y ,marker='*',color="red")
plt.title("EXAMPLE")
plt.xlabel("X")
plt.ylabel("Y")
plt.show
```

```python
print(type(plt.title))
del plt.title
```

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]
plt.bar(x,y)
plt.bar(x,y ,color="pink")
plt.title("EXAMPLE")
plt.xlabel("Color")
plt.ylabel("Count")
plt.show
```

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]
plt.pie(y,labels=x, autopct="%1.f%%", colors=["red","green","blue","orange"])
plt.show()
```
